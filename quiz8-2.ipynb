```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

vti = pd.read_csv('VTI_12_31_2019.csv', index_col=0)
vxf = pd.read_csv('VXF_12_31_2019.csv', index_col=0)
vb = pd.read_csv('VB_12_31_2019.csv', index_col=0)
vo = pd.read_csv('VO_12_31_2019.csv', index_col=0)
voo = pd.read_csv('VOO_12_31_2019.csv', index_col=0)
vti['etf'] = 'VTI'
vxf['etf'] = 'VXF'
vb['etf'] = 'VB'
vo['etf'] = 'VO'
voo['etf'] = 'VOO'

vti['marketValue'] = vti['marketValue']*.15
vxf['marketValue'] = vxf['marketValue']*.30
vb['marketValue'] = vb['marketValue']*.30
vo['marketValue'] = vo['marketValue']*.15
voo['marketValue'] = voo['marketValue']*.10

df = pd.concat([vti, vb, vxf, vo, voo], axis=0)
df.sort_values('marketValue', ascending=False, inplace=True, ignore_index=True)
df['%'] = df['marketValue']/df['marketValue'].sum()*100.0

print(df.iloc[0:50])
print(df.iloc[50:500])
print(df.iloc[500:])

print()
print('Top 50: {:.8f}%'.format(100.0*df.iloc[0:50]['marketValue'].sum()/df['marketValue'].sum()))
print('Next 450: {:.8f}%'.format(100.0*df.iloc[50:500]['marketValue'].sum()/df['marketValue'].sum()))
print('Remaining: {:.8f}%'.format(100.0*df.iloc[500:]['marketValue'].sum()/df['marketValue'].sum()))
```
